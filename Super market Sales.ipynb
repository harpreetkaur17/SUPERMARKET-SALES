```python
products=data['Product line'].unique()
x=1
plt.figure(figsize=[16,16])
for i in products:
    ax=plt.subplot(3,2,x)
    pvt1=pd.pivot_table(data[data['Product line']==i],columns='City',index='WeekDay',aggfunc='sum',values='gross income')
    pvt1.plot(title=i,ax=ax)
    plt.ylabel("gross income")
    plt.legend()
    x=x+1
```

## Irrespective of branches

```python
pvt=pd.pivot_table(data,columns='Product line',index='Month',aggfunc='sum',values='gross income')
pvt.plot()
```

### As we know qunatity and gross income are directly proportion so we can assume similar kind of graph for quantity 
### Below is the graph for the same 

```python
pvt=pd.pivot_table(data,columns='Product line',index='Month',aggfunc='sum',values='Quantity')

pvt.plot()
plt.xlabel("Month")
plt.ylabel("gross income")
plt.legend()
plt.show()
```


### gross income and qunatity of sale of Food and beverages ,Fashion accessories are high during second month and lowest in third month as compare other quality  irrespective of branches
### Sports and travel , Home and lifestyle are highest in first month
### As we see dip in gross income of all the product lines except the two we can predict that overall income will also be low for all branches during the second month , which can be seen from below graph

```python
plt.figure(figsize=[8,5])

Mandalay_report=data[data['City']=="Mandalay"].groupby('Month')['gross income'].sum()
Mandalay_report.plot(color='red',label="Mandalay")
Yangon_report=data[data['City']=="Yangon"].groupby('Month')['gross income'].sum()
Yangon_report.plot(color='green',label="Yangon")
Naypyitaw_report=data[data['City']=="Naypyitaw"].groupby('Month')['gross income'].sum()
Naypyitaw_report.plot(color='blue',label="Naypyitaw")
plt.xlabel("Month")
plt.ylabel("gross income")
plt.legend()
plt.show()
```

## Monthy Gross income variation with respect to different cities and Product lines

```python
plt.figure(figsize=[16,4])
x=1
for i in city:
    ax1=plt.subplot(1,3,x)
    pvt=pd.pivot_table(data[data['City']==i],columns='Product line',index='Month',aggfunc='sum',values='gross income')
    pvt.plot(use_index=True,title=i,ax=ax1)
    plt.xlabel("Month")
    plt.ylabel("gross income")
    plt.legend()
    x=x+1
```
